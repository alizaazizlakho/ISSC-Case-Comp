```python
#Generating a new dataframe with only language information
profile_data = profile_data.set_index("Category")
lang_data = profile_data.drop(["Neighbourhood Information", "Population", "Families, households and marital status", "Income", "Immigration and citizenship", "Visible minority", "Ethnic origin", "Aboriginal peoples", "Housing", "Education", "Language of work", "Labour", "Journey to work", "Mobility"])
```

```python
lang_data.describe
```

```python
lang_data = lang_data.set_index("Topic")
```

```python
#Generating a more refined dataframe with only language information
lang_data_refined = lang_data.drop(["Knowledge of official languages", "First official language spoken", "Mother tongue", "Language spoken most often at home", "Other language spoken regularly at home"])
```

```python
lang_data_refined = lang_data_refined.set_index("Characteristic")
```

```python
lang_data_refined.head()
```

```python
lang_data_refined = lang_data_refined.drop(columns=["_id", "Data Source"])
```
